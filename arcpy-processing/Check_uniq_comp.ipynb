```python
def feature_class_to_pandas_data_frame(fc, flds):
    """
    Load data into a Pandas Data Frame for subsequent analysis.
    :param feature_class: Input ArcGIS Feature Class.
    :param field_list: Fields for input.
    :return: Pandas DataFrame object.
    """
    return pd.DataFrame(
        arcpy.da.FeatureClassToNumPyArray(
            in_table=fc,
            field_names=[fld.name for fld in arcpy.ListFields(fc) if fld.name != arcpy.Describe(fc).shapeFieldName],
            skip_nulls=False,
            null_value=-99999
        )
    )
```

```python
df = feature_class_to_pandas_data_frame(fc, flds)
```
