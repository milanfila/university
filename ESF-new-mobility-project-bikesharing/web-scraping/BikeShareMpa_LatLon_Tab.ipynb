```python
import requests
import json
import pandas as pd
import geopandas as geopd
from shapely.geometry import Point
```

```python
def bikesharemap():

    headers = {
        'Connection': 'keep-alive',
        'sec-ch-ua': '" Not A;Brand";v="99", "Chromium";v="98", "Google Chrome";v="98"',
        'Accept': '*/*',
        'sec-ch-ua-mobile': '?0',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36',
        'sec-ch-ua-platform': '"Windows"',
        'Origin': 'https://bikesharemap.com',
        'Sec-Fetch-Site': 'cross-site',
        'Sec-Fetch-Mode': 'cors',
        'Sec-Fetch-Dest': 'empty',
        'Referer': 'https://bikesharemap.com/',
        'Accept-Language': 'cs-CZ,cs;q=0.9',
    }

    response = requests.get('https://julie.geog.ucl.ac.uk/~ollie/bikesapi/systems_js.php', headers=headers)
    return response.json()
```

```python
data_map = bikesharemap()
```

```python
df = pd.DataFrame.from_dict(data_map, orient="index")
```

```python
geometry = [Point(LonLat) for LonLat in zip(df.lon, df.lat)]
crs = {'init': 'epsg:4326'}
gdf = geopd.GeoDataFrame(df, crs=crs, geometry=geometry)
```
